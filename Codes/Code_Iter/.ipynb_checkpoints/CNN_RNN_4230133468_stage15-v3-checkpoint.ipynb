```python
from numpy.random import seed
seed(42)
from tensorflow import set_random_seed
set_random_seed(42)
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.python.client import device_lib
import matplotlib.pyplot as plt
import numpy as np
from numpy import concatenate
from numpy import sqrt
import math
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, SimpleRNN
from tensorflow.keras.layers import LSTM, Dropout
from keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
FileName = '../Data/'
df = pd.read_csv(FileName)
```

```python
config = tf.ConfigProto(device_count = {'GPU':1, 'CPU':32})
sess = tf.Session(config=config)
keras.backend.set_session(sess)
print(device_lib.list_local_devices())
```

## Set all the stages to 15 ? ##

```python
stage_no = 15
```

```python
df = df[(df.stage_number == stage_no)]
```

## fr_conc_2 is the only frac_cont_# with non-empty elements ## 
## what is the meaning of _# numbers ? ##

```python
df = df[['date_time','prop_conc_1','bh_prop_conc','fr_conc_2','slurry_rate','wellhead_pressure_1']]
```

```python
df['date_time']=pd.to_datetime(df.date_time.iloc[:])
```

```python
df.head()
```

```python
t = (df.date_time-df.date_time.iloc[0]).dt.total_seconds()
df.loc[:,'new_time'] = t
df.head()
```

```python
df1 = df[(df.slurry_rate > 0 )]
df1 = df[(df.prop_conc_1>0)]
```

```python
df1.new_time = df.new_time-min(df.new_time)
```

```python
df1.prop_conc_1 =df1.prop_conc_1 - df1.prop_conc_1.mean()
df1.bh_prop_conc = df1.bh_prop_conc - df1.bh_prop_conc.mean()
df1.fr_conc_2 = df1.fr_conc_2 - df1.fr_conc_2.mean()
df1.slurry_rate = df1.slurry_rate - df1.slurry_rate.mean()
df1.wellhead_pressure_1 = df1.wellhead_pressure_1 - df1.wellhead_pressure_1.mean()
```

```python
df1
```

```python
fig=plt.figure(figsize=(14, 8), dpi= 80, facecolor='w', edgecolor='k')

plt.subplot(511)
plt.plot(df1.new_time,df1.prop_conc_1)
plt.ylabel('prop_conc')

plt.subplot(512)
plt.plot(df1.new_time,df1.bh_prop_conc)
plt.ylabel('bh_prop_conc')

plt.subplot(513)
plt.plot(df1.new_time,df1.fr_conc_2)
plt.ylabel('fr_conc')

plt.subplot(514)
plt.plot(df1.new_time,df1.slurry_rate)
plt.ylabel('Slurry Rate')

plt.subplot(515)
plt.plot(df1.new_time,df1.wellhead_pressure_1)
plt.ylabel('Wellhead Pressure')
```

```python
Nrow = df1.shape[0]
Sample_size = 120
Nrow1 = Nrow - Sample_size
data1 = df1[['prop_conc_1','bh_prop_conc','fr_conc_2','slurry_rate']]
data2 = df1['wellhead_pressure_1']
X = np.zeros((Nrow1, Sample_size,4))
Y = np.zeros((Nrow1,1))
time_n = np.zeros((Nrow1,1))
for i in range(0, Nrow1):
    X[i,:,:] = data1.iloc[i:(i+Sample_size),:]
    Y[i] = data2.iloc[(i+Sample_size-1)]
    time_n[i] = df1.new_time.iloc[i+Sample_size]
df2 = pd.DataFrame()
for i in range(0, Sample_size):
    col_name1 = "prop_conc(t-" + str(Sample_size-1-i) + ")"
    df2[col_name1] = pd.Series(X[:,i,0])
for i in range(0, Sample_size):
    col_name2 = "bp_prop_conc(t-" + str(Sample_size-1-i) + ")"
    df2[col_name2] = pd.Series(X[:,i,1])
for i in range(0, Sample_size):
    col_name3 = "fr_conc_2(t-" + str(Sample_size-1-i) + ")"
    df2[col_name3] = pd.Series(X[:,i,2])
for i in range(0, Sample_size):
    col_name4 = "slurry_rate(t-" + str(Sample_size-1-i) + ")"
    df2[col_name4] = pd.Series(X[:,i,3])
# Y = df1[['wellhead_pressure_1']]
```

```python
#Y.reset_index(drop=True, inplace=True)
```

```python
dataset = df2.values
N_feature = dataset.shape[1] - 1
dataset = dataset.astype('float32')

# normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
scaler = scaler.fit(dataset)
dataset = scaler.transform(dataset)
# split into train and test sets
train_size = 6000
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:,:]
print(len(train), len(test))
```

```python
trainX = train[:,0:N_feature]
trainY = train[:,N_feature]
testX = test[:,0:N_feature]
testY = test[:,N_feature]

#reshape input to be [samples, time steps, features]
trainX = np.reshape(trainX,(trainX.shape[0], -1, 4))
testX = np.reshape(testX, (testX.shape[0], -1, 4))
trainX_sub, X_valid, trainY_sub, y_valid = train_test_split(trainX, trainY, test_size=0.2)
```

```python
model = Sequential()
model.add(Conv1D(10,3,input_shape=(Sample_size,4), activation = 'relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(10,3, activation = 'relu'))
model.add(MaxPooling1D(pool_size=2))
#model.add(Conv1D(10,3, activation = 'relu'))
#model.add(MaxPooling1D(pool_size=2))
#model.add(Conv1D(10,3, activation = 'relu'))
#model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(10,3, activation = 'relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(SimpleRNN(10, return_sequences = True))
model.add(Dropout(0.2))
model.add(SimpleRNN(10))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1)
history = model.fit(trainX_sub, trainY_sub, validation_data=(X_valid, y_valid), epochs=100, verbose=0, callbacks=[es])
# evaluate the model
train_acc = model.evaluate(trainX_sub, trainY_sub, verbose=0)
test_acc = model.evaluate(X_valid, y_valid, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```

```python
# plot training history
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
# evaluate the model
#model.fit(trainX_sub, trainY_sub, epochs=100, batch_size=200, validation_data=(X_valid, y_valid), verbose=1)
```

```python
# make predictions
from numpy import concatenate
from numpy import sqrt
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
fig=plt.figure(figsize=(14, 8), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(time_n[0:len(train)],trainY,'r--')
plt.plot(time_n[0:len(train)],trainPredict,'r')
plt.plot(time_n[len(train):],testY,'g--')
plt.plot(time_n[len(train):],testPredict,'g')
```

```python
# invert predictions
testX = testX.reshape((testX.shape[0], -1))
trainX = trainX.reshape((trainX.shape[0], -1))
```

```python
# invert scaling for forecast
inv_testPredict = concatenate((testX, testPredict), axis=1)
inv_testPredict = scaler.inverse_transform(inv_testPredict)
inv_testPredict = inv_testPredict[:,-1]
# invert scaling for actual
testY = testY.reshape((len(testY), 1))
inv_testY = concatenate((testX, testY), axis=1)
inv_testY = scaler.inverse_transform(inv_testY)
inv_testY = inv_testY[:,-1]

# calculate RMSE
rmse = sqrt(mean_squared_error(inv_testY, inv_testPredict))
print('Test RMSE: %.3f' % rmse)
# invert train data


# invert scaling for forecast
inv_trainPredict = concatenate((trainX, trainPredict), axis=1)
inv_trainPredict = scaler.inverse_transform(inv_trainPredict)
inv_trainPredict = inv_trainPredict[:,-1]
# invert scaling for actual
trainY = trainY.reshape((len(trainY), 1))
inv_trainY = concatenate((trainX, trainY), axis=1)
inv_trainY = scaler.inverse_transform(inv_trainY)
inv_trainY = inv_trainY[:,-1]

# calculate RMSE
rmse = sqrt(mean_squared_error(inv_trainY, inv_trainPredict))
print('Train RMSE: %.3f' % rmse)
fig=plt.figure(figsize=(14, 14), dpi= 80, facecolor='w', edgecolor='k')
plt.subplot(211)

plt.plot(time_n[0:len(train)],inv_trainY,'r--', label = "Traing data: original")
plt.plot(time_n[0:len(train)],inv_trainPredict,'r', label = "Training data: fitted")
plt.xlabel('time',fontsize=20)
plt.ylabel('Wellhead pressure',fontsize=20)
plt.tick_params(labelsize=20)
plt.legend(loc='lower right', prop={'size': 20})

plt.subplot(212)
plt.plot(time_n[len(train):],inv_testY,'g--', label = 'Test data: original')
plt.plot(time_n[len(train):],inv_testPredict, 'g', label = 'Test data: predicted')
plt.xlabel('time',fontsize=20)
plt.ylabel('Wellhead pressure',fontsize=20)
plt.tick_params(labelsize=20)
plt.legend(loc='lower right', prop={'size': 20})
```
